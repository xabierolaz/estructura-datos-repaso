## Stack

Regla: LIFO.

```python
stack = []
for value in [4, 9, 2]:
    stack.append(value)

x = stack.pop()
y = stack[-1]
print("x =", x, "y =", y)
assert x == 2
assert y == 9
```

## Queue

Regla: FIFO.

```python
queue = []
for value in ["A", "B", "C"]:
    queue.append(value)

x = queue.pop(0)
y = queue[0]
print("x =", x, "y =", y)
assert x == "A"
assert y == "B"
```
