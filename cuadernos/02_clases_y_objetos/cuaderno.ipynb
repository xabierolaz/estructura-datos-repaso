## `__init__`, `self`, atributos y metodos

Regla del bloque:

- `__init__` define como nace el objeto
- `self` significa este objeto concreto
- atributo = dato
- metodo = comportamiento sobre ese estado

```python
class Counter:
    def __init__(self):
        self.total = 0

    def add(self, value):
        self.total += value


c = Counter()
c.add(3)
c.add(7)
assert c.total == 10
```

## Funcion frente a metodo

Esto debe quedar limpio antes de seguir a `Stack` y `Queue`.

```python
def suma(a, b):
    return a + b


class Caja:
    def __init__(self, valor):
        self.valor = valor

    def duplicar(self):
        self.valor *= 2


caja = Caja(5)
caja.duplicar()
assert suma(2, 4) == 6
assert caja.valor == 10
```
